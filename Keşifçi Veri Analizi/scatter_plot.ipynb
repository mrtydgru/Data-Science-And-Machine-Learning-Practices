## Scatterplot Matrisi

```python
import seaborn as sns;
iris = sns.load_dataset("iris")
df = iris.copy()
df.head()
```

```python
df.dtypes
```

```python
df.shape
```

```python
sns.pairplot(data=df);
```

```python
sns.pairplot(data=df, hue="species");
```

```python
sns.pairplot(data=df, hue="species", markers=["o","s","D"]);
```

```python
sns.pairplot(data=df, kind="reg", hue="species");
```
